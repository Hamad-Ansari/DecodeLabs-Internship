```python
section(
    "Task 2: Data Cleaning & Preprocessing",
    "Fix data types, remove duplicates, handle missing values, and create new analysis features.",
    "🧹"
)

data = df.copy()
raw_rows = len(data)

# ------------------------------------------------------------
# Standardize column names
# ------------------------------------------------------------
canonical_columns = {
    "orderid": "OrderID",
    "date": "Date",
    "customerid": "CustomerID",
    "product": "Product",
    "quantity": "Quantity",
    "unitprice": "UnitPrice",
    "shippingaddress": "ShippingAddress",
    "paymentmethod": "PaymentMethod",
    "orderstatus": "OrderStatus",
    "trackingnumber": "TrackingNumber",
    "itemsincart": "ItemsInCart",
    "couponcode": "CouponCode",
    "referralsource": "ReferralSource",
    "totalprice": "TotalPrice"
}

data.columns = (
    data.columns.astype(str)
    .str.strip()
    .str.replace(" ", "", regex=False)
    .str.replace("_", "", regex=False)
    .str.replace("-", "", regex=False)
)

data = data.rename(columns={col: canonical_columns.get(col.lower(), col) for col in data.columns})

# ------------------------------------------------------------
# Remove duplicates
# ------------------------------------------------------------
duplicate_rows = data.duplicated().sum()
data = data.drop_duplicates().reset_index(drop=True)

# ------------------------------------------------------------
# Convert Date column
# ------------------------------------------------------------
if "Date" in data.columns:
    data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
    
    if data["Date"].isna().any() and data["Date"].notna().any():
        data = data.dropna(subset=["Date"]).reset_index(drop=True)

# ------------------------------------------------------------
# Convert numeric columns
# ------------------------------------------------------------
base_numeric_cols = [col for col in ["Quantity", "UnitPrice", "ItemsInCart"] if col in data.columns]

for col in base_numeric_cols:
    data[col] = pd.to_numeric(data[col], errors="coerce")
    data.loc[data[col] < 0, col] = np.nan
    
    fill_value = data[col].median()
    if pd.isna(fill_value):
        fill_value = 0
        
    data[col] = data[col].fillna(fill_value)
    
    if col in ["Quantity", "ItemsInCart"]:
        data[col] = data[col].round().astype(int)

if "TotalPrice" in data.columns:
    data["TotalPrice"] = pd.to_numeric(data["TotalPrice"], errors="coerce")
    data.loc[data["TotalPrice"] < 0, "TotalPrice"] = np.nan

# ------------------------------------------------------------
# Create calculated gross value
# ------------------------------------------------------------
if {"Quantity", "UnitPrice"}.issubset(data.columns):
    data["CalculatedGross"] = data["Quantity"].astype(float) * data["UnitPrice"].astype(float)
    
    if "TotalPrice" not in data.columns:
        data["TotalPrice"] = data["CalculatedGross"]
    else:
        data["TotalPrice"] = data["TotalPrice"].fillna(data["CalculatedGross"])
        data["EstimatedDiscountOrAdjustment"] = data["CalculatedGross"] - data["TotalPrice"]

if "TotalPrice" in data.columns:
    fill_total = data["TotalPrice"].median()
    if pd.isna(fill_total):
        fill_total = 0
    data["TotalPrice"] = data["TotalPrice"].fillna(fill_total)

# ------------------------------------------------------------
# Clean text columns
# ------------------------------------------------------------
text_cols = data.select_dtypes(include=["object", "category"]).columns.tolist()

for col in text_cols:
    data[col] = data[col].astype(str).str.strip()
    data[col] = data[col].replace(
        ["", "nan", "NaN", "None", "none", "NULL", "null", "N/A", "n/a"],
        np.nan
    )

if "CouponCode" in data.columns:
    data["CouponCode"] = data["CouponCode"].fillna("No Coupon")
    data["CouponCode"] = data["CouponCode"].astype(str).str.strip()
    data["CouponCode"] = data["CouponCode"].replace(
        ["NO COUPON", "No coupon", "None", "nan", "Unknown"],
        "No Coupon"
    )

for col in ["Product", "ShippingAddress", "PaymentMethod", "OrderStatus", "TrackingNumber", "ReferralSource", "CustomerID", "OrderID"]:
    if col in data.columns:
        data[col] = data[col].fillna("Unknown")

for col in ["Product", "PaymentMethod", "OrderStatus", "ReferralSource"]:
    if col in data.columns:
        data[col] = data[col].astype(str).str.strip().str.title()

# ------------------------------------------------------------
# Feature engineering
# ------------------------------------------------------------
if "Date" in data.columns:
    data["Year"] = data["Date"].dt.year
    data["Month"] = data["Date"].dt.month
    data["MonthName"] = data["Date"].dt.month_name()
    data["Quarter"] = data["Date"].dt.quarter
    data["DayName"] = data["Date"].dt.day_name()
    data["DayOfWeek"] = data["Date"].dt.dayofweek
    data["YearMonth"] = data["Date"].dt.to_period("M").astype(str)
    data["MonthDate"] = data["Date"].dt.to_period("M").dt.to_timestamp()

if "CouponCode" in data.columns:
    no_coupon_values = {
        "no coupon",
        "no_coupon",
        "none",
        "nan",
        "unknown",
        "not available",
        "not used",
        ""
    }
    data["HasCoupon"] = ~data["CouponCode"].astype(str).str.strip().str.lower().isin(no_coupon_values)
else:
    data["HasCoupon"] = False

if "OrderStatus" in data.columns:
    data["IsCancelled"] = data["OrderStatus"].astype(str).str.lower().eq("cancelled").astype(int)
    data["IsReturned"] = data["OrderStatus"].astype(str).str.lower().eq("returned").astype(int)
    data["IsDelivered"] = data["OrderStatus"].astype(str).str.lower().eq("delivered").astype(int)

if {"OrderStatus", "TotalPrice"}.issubset(data.columns):
    delivered_mask = data["OrderStatus"].astype(str).str.lower().eq("delivered")
    data["NetRevenue"] = np.where(delivered_mask, data["TotalPrice"], 0)
else:
    if "TotalPrice" in data.columns:
        data["NetRevenue"] = data["TotalPrice"]

# ------------------------------------------------------------
# Cleaning summary
# ------------------------------------------------------------
new_columns = [col for col in data.columns if col not in df.columns]

cleaning_summary = pd.DataFrame({
    "Step": [
        "Rows before cleaning",
        "Duplicate rows removed",
        "Rows after cleaning",
        "Remaining missing cells",
        "New engineered columns"
    ],
    "Value": [
        raw_rows,
        duplicate_rows,
        len(data),
        int(data.isna().sum().sum()),
        ", ".join(new_columns) if new_columns else "None"
    ]
})

display(style_table(cleaning_summary, "Cleaning Summary"))
display(data.head())
```

```python
section(
    "Task 5: Predictive Model",
    "Build a simple model to predict whether an order may be cancelled.",
    "🚀"
)

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report, confusion_matrix

if "OrderStatus" not in data.columns:
    raise ValueError("OrderStatus column is required for cancellation prediction.")

ml_df = data.copy()
ml_df["CancelledFlag"] = ml_df["OrderStatus"].astype(str).str.lower().eq("cancelled").astype(int)

feature_candidates = [
    "Product",
    "Quantity",
    "UnitPrice",
    "ItemsInCart",
    "PaymentMethod",
    "ReferralSource",
    "HasCoupon",
    "Month",
    "DayOfWeek"
]

feature_cols = [col for col in feature_candidates if col in ml_df.columns]

if not feature_cols:
    raise ValueError("No usable feature columns found for modeling.")

X = ml_df[feature_cols].copy()
y = ml_df["CancelledFlag"]

if y.nunique() < 2:
    print("Model cannot be trained because the target has only one class.")
else:
    numeric_features = X.select_dtypes(include=np.number).columns.tolist()
    categorical_features = [col for col in X.columns if col not in numeric_features]
    
    numeric_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median"))
    ])
    
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ])
    
    transformers = []
    
    if numeric_features:
        transformers.append(("num", numeric_transformer, numeric_features))
    
    if categorical_features:
        transformers.append(("cat", categorical_transformer, categorical_features))
    
    preprocessor = ColumnTransformer(transformers=transformers)
    
    model = RandomForestClassifier(
        n_estimators=350,
        random_state=42,
        class_weight="balanced",
        min_samples_leaf=2
    )
    
    clf = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("model", model)
    ])
    
    stratify_target = y if y.value_counts().min() >= 2 else None
    
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.25,
        random_state=42,
        stratify=stratify_target
    )
    
    clf.fit(X_train, y_train)
    
    y_pred = clf.predict(X_test)
    
    if hasattr(clf, "predict_proba"):
        y_proba = clf.predict_proba(X_test)[:, 1]
    else:
        y_proba = None
    
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    if y_proba is not None and y_test.nunique() == 2:
        auc = roc_auc_score(y_test, y_proba)
    else:
        auc = np.nan
    
    kpi_cards([
        ("🎯", "Accuracy", f"{accuracy:.3f}", "Overall correct predictions"),
        ("⚖️", "F1 Score", f"{f1:.3f}", "Balance between precision and recall"),
        ("📈", "ROC AUC", f"{auc:.3f}" if not pd.isna(auc) else "N/A", "Model ranking quality"),
        ("🧪", "Test Records", f"{len(X_test):,}", "Records used for evaluation")
    ])
    
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=["Not Cancelled", "Cancelled"], zero_division=0))
    
    # --------------------------------------------------------
    # Confusion Matrix
    # --------------------------------------------------------
    cm = confusion_matrix(y_test, y_pred)
    
    fig, ax = plt.subplots(figsize=(7, 5), facecolor=COLORS["cream"])
    
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap=GREEN_CMAP,
        cbar=False,
        xticklabels=["Not Cancelled", "Cancelled"],
        yticklabels=["Not Cancelled", "Cancelled"],
        linewidths=0.5,
        linecolor=COLORS["cream"],
        ax=ax
    )
    
    ax.set_title("Confusion Matrix", fontweight="bold", color=COLORS["dark"], pad=14)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")
    
    plt.tight_layout()
    plt.show()
    
    # --------------------------------------------------------
    # Feature Importance
    # --------------------------------------------------------
    try:
        feature_names = clf.named_steps["preprocess"].get_feature_names_out()
        importances = clf.named_steps["model"].feature_importances_
        
        feature_importance = (
            pd.DataFrame({
                "Feature": feature_names,
                "Importance": importances
            })
            .sort_values("Importance", ascending=False)
            .head(15)
        )
        
        feature_importance["Feature"] = (
            feature_importance["Feature"]
            .str.replace("num__", "", regex=False)
            .str.replace("cat__", "", regex=False)
        )
        
        fig, ax = plt.subplots(figsize=(12, 7), facecolor=COLORS["cream"])
        
        sns.barplot(
            data=feature_importance,
            x="Importance",
            y="Feature",
            color=COLORS["sage"],
            ax=ax
        )
        
        polish(ax, "Top Features Influencing Cancellation Prediction", "Importance", "Feature")
        plt.tight_layout()
        plt.show()
        
        display(style_table(feature_importance, "Top Model Features"))
        
    except Exception as e:
        print("Feature importance could not be displayed:", e)
```

```python
section(
    "Insight-Based Business Recommendations",
    "Actionable decisions based on the analysis.",
    "💡"
)

insights = []

# Product insight
if {"Product", "TotalPrice"}.issubset(analysis_df.columns):
    product_revenue = analysis_df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)
    
    if not product_revenue.empty:
        best_product = product_revenue.index[0]
        insights.append((
            "🛍️ Product Strategy",
            f"<b>{best_product}</b> generated the highest revenue. Maintain strong inventory, promote bundles, and use this category in campaigns."
        ))

# Referral insight
if {"ReferralSource", "TotalPrice"}.issubset(analysis_df.columns):
    referral_revenue = analysis_df.groupby("ReferralSource")["TotalPrice"].sum().sort_values(ascending=False)
    
    if not referral_revenue.empty:
        best_referral = referral_revenue.index[0]
        insights.append((
            "📣 Marketing Strategy",
            f"<b>{best_referral}</b> is the strongest revenue-driving referral source. Consider increasing marketing budget on this channel."
        ))

# Coupon insight
if {"HasCoupon", "TotalPrice"}.issubset(data.columns):
    coupon_aov = data.groupby("HasCoupon")["TotalPrice"].mean()
    
    if True in coupon_aov.index and False in coupon_aov.index:
        delta = coupon_aov.loc[True] - coupon_aov.loc[False]
        
        if delta > 0:
            msg = f"Coupon orders have a higher AOV by approximately <b>{money(delta)}</b>. Coupons may be helping increase basket size."
        else:
            msg = f"Coupon orders have a lower AOV by approximately <b>{money(abs(delta))}</b>. Review discount strategy to protect margins."
        
        insights.append(("🏷️ Coupon Strategy", msg))

# Cancellation insight
if {"Product", "IsCancelled"}.issubset(data.columns):
    cancel_by_product = (
        data
        .groupby("Product")
        .agg(
            CancelRate=("IsCancelled", "mean"),
            Orders=("IsCancelled", "size")
        )
        .query("Orders >= 5")
        .sort_values("CancelRate", ascending=False)
    )
    
    if not cancel_by_product.empty:
        risky_product = cancel_by_product.index[0]
        risky_rate = cancel_by_product.iloc[0]["CancelRate"] * 100
        
        insights.append((
            "❌ Cancellation Reduction",
            f"<b>{risky_product}</b> has one of the highest cancellation rates at <b>{risky_rate:.1f}%</b>. Check pricing, stock availability, shipping delays, or product expectations."
        ))

# Payment method insight
if "PaymentMethod" in data.columns:
    top_payment = data["PaymentMethod"].value_counts().idxmax()
    
    insights.append((
        "💳 Payment Experience",
        f"<b>{top_payment}</b> is the most used payment method. Ensure this method remains fast, secure, and error-free."
    ))

# Display insight cards
if insights:
    html = '<div class="insight-grid">'
    
    for title, body in insights:
        html += f"""
        <div class="insight-card">
            <h3>{title}</h3>
            <p style="font-size:15px; line-height:1.7;">{body}</p>
        </div>
        """
    
    html += "</div>"
    display(HTML(html))
else:
    display(Markdown("No insights could be generated from the available columns."))
```
